```python
import pandas as pd                   #for data handling (espically with data_frames)
from sklearn import preprocessing     #for labeling cuisines (apply numerical labels or ids on cuisines )
from sklearn.feature_extraction.text import TfidfVectorizer #to convert text data to numericals 
                                                             #without losing any property or parameter

from sklearn.model_selection import train_test_split  #for spilitting data set into test data set and train data set 
from sklearn.metrics import accuracy_score, confusion_matrix  #for confusion matrix and accuracy score
from sklearn.svm import SVC
from sklearn import svm #our training algorithm 
from sklearn.model_selection import GridSearchCV # optional(for hyper parameter tuning)
import os
print("done")
```

```python
#retrieving data from .json files
df_train = pd.read_json("./input/트레인.json")
df_test = pd.read_json("./input/te23테스트.json")
df_test
```

```python
#재료 데이터 뽑기
train = pd.DataFrame({'ingredients':df_train['ingredients']})
train2 = train.join(df_train['ingredients'].apply(lambda x: pd.Series(x)).stack().reset_index(1,name='ingredients_new').drop('level_1', axis=1))
train3 = pd.DataFrame({'ingredients':train2['ingredients_new']})
print(train2)
print(train3)

train3.to_csv('submit5.csv', index = False)
print('done')
```

```python
tra = pd.DataFrame({'cuisine':df_test['id'], 'ingredients':df_test['ingredients']})
train4 = tra.join(df_test['ingredients'].apply(lambda x: pd.Series(x)))
print(train4)

train5 = train4.drop(['ingredients'], axis=1)
print(train5)

train5.to_excel('test.xlsx', index = False)
train5.to_csv('test.csv', index = False)
print('done')
```
